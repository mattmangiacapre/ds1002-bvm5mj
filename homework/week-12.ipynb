## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

plt.pie(slices, labels = ["US","UK","MX","CA","JP"])
plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
plt.hist(n, bins=30)
plt.title("Histogram")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
z = np.polyfit(x, y, 3) # devloping math for regression
p = np.poly1d(z) # graphic implications for regression
plt.plot(x, p(x)) # plotting regression
plt.scatter(x, y)
plt.title("Scatter with Trendline") # labeling
plt.show()
```
